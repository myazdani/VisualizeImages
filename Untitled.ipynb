```python
from SkiCaffe.skicaffe import SkiCaffe
import os
```

```python
caffe_root = '/usr/local/src/caffe/caffe-master/'
```

```python
DLmodel = SkiCaffe('/usr/local/src/caffe/caffe-master/')
```

```python
model_prototxt = caffe_root + 'models/bvlc_googlenet/deploy.prototxt'
model_trained = caffe_root + 'models/bvlc_googlenet/bvlc_googlenet.caffemodel'
```

```python
DLmodel.fit(model_prototxt_path = model_prototxt, model_trained_path = model_trained)
print 'Number of layers:', len(DLmodel.layer_sizes)
DLmodel.layer_sizes
```

```python
src_path, image_type = "./data/Rothko_images/", ".jpg"
image_paths = []  
for root, dirs, files in os.walk(src_path):
    image_paths.extend([os.path.join(root, f) for f in files if f.endswith(image_type)])
```

```python
len(image_paths)
```

```python
image_features = DLmodel.transform(image_paths = image_paths, layer_name = 'pool5/7x7_s1', return_type = 'pandasDF')
image_features.head()
```

```python
image_features.to_csv("googlenet_features.csv", index = False)
```
